```python
#Importing packages
import numpy as np
import pandas as pd

import warnings
warnings.filterwarnings("ignore")

#Importing visualization packages
import matplotlib.pyplot as plt
%matplotlib inline

from matplotlib.ticker import ScalarFormatter

import seaborn as sns
```

```python
#Importing the data for years 1986-2006 as panda dataframe
df_pop1 = pd.read_csv('1710011801-noSymbol.csv', skiprows=6, nrows=13)

df_pop1
```

```python
#Cleaning the dataframe:

#Removing the superscripts from the "Population characteristics" column
df_pop1['Population characteristics'] = df_pop1['Population characteristics'].str.replace('\d+', '')

df_pop1
```

```python
#Importing the data for years 1996-2016 as panda dataframe
df_pop2 = pd.read_csv('3810004701-noSymbol.csv', skiprows=5, nrows=19)

df_pop2
```

```python
#Cleaning the dataframe:

#Removing the superscripts from the "Population characteristics" column
df_pop2['Population characteristics'] = df_pop2['Population characteristics'].str.replace('\d+', '')

#Dropping the first column since it does not provide additional information for the analysis
df_pop2 = df_pop2.drop('Geography4', axis=1)

#Dropping rows with NaN values since they do not provide additional information for the analysis
df_pop2 = df_pop2.dropna()

df_pop2
```

```python
#Resetting the index on df_pop2 after its row removals, so the dataframes can be joined correctly:
df_pop2 = df_pop2.reset_index(drop=True)
df_pop2
```

```python
#Adding the last two columns of df_pop2 (2011 & 2016) to df_pop1 to create dataframe for years 1986-2016:

df_pop1['2011'] = df_pop2['2011']
df_pop1['2016'] = df_pop2['2016']

df_pop1
```

```python
df_pop1.info()
```

```python
#Converting all "object" types to "float" numbers:

df_pop1['1986'] = df_pop1['1986'].str.replace(',', "").astype(float)
df_pop1['1991'] = df_pop1['1991'].str.replace(',', "").astype(float)
df_pop1['1996'] = df_pop1['1996'].str.replace(',', "").astype(float)
df_pop1['2001'] = df_pop1['2001'].str.replace(',', "").astype(float)
df_pop1['2006'] = df_pop1['2006'].str.replace(',', "").astype(float)
df_pop1['2011'] = df_pop1['2011'].str.replace(',', "").astype(float)
df_pop1['2016'] = df_pop1['2016'].str.replace(',', "").astype(float)

df_pop1
```

```python
#Setting the index to the first column & transposing the dataframe:

df_pop1.set_index('Population characteristics', inplace=True)

df_pop1 = df_pop1.transpose()

df_pop1
```

```python
df_pop1.columns
```

```python
#Investigating changes in urban/rural dwellings over the years 1986-2016:

#Calculating the urban/rural dwellings as a share of the total dwellings:
df_pop1['Percent urban dwellings'] = df_pop1['Urban private dwellings  '] / df_pop1['Total private dwellings '] * 100
df_pop1['Percent rural dwellings'] = df_pop1['Rural private dwellings  '] / df_pop1['Total private dwellings '] * 100

df_pop1
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12,4))

dwellings_plot = df_pop1[['Urban private dwellings  ', 'Rural private dwellings  ']].plot(kind='bar', legend=True, ax=ax1)
dwellings_plot.set_xlabel("Year")
dwellings_plot.set_ylabel("Number of Dwellings")
dwellings_plot.ticklabel_format(style='plain', axis='y')

perc_dwellings_plot = df_pop1[['Percent urban dwellings', 'Percent rural dwellings']].plot(kind='bar', legend=True, ax=ax2)
perc_dwellings_plot.set_xlabel("Year")
perc_dwellings_plot.set_ylabel("Percentage of Dwellings")

plt.tight_layout();
```

```python
#As seen, the number of urban private dwellings has had a significant increase over the years,
#jumping from 7.1 million dwellings in 1986 to approximately 11.5 million dwellings in 2016.
#Between 1986 and 2016, the number of rural private dwellings is almost constant with a slight general increase.

#Even though the number of the private dwellings show an increase between 1986 and 2016, 
#their share of the total number of dwellings (percentage of urban/rural dwellings) remains without significant change.
```
